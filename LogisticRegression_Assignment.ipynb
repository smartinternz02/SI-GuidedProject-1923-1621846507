```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("bikebuyer1.csv")
```

```python
dataset
```

```python
dataset.isnull().any()
```

```python
dataset.isnull().sum()
```

```python
dataset["Marital Status"].fillna(dataset["Marital Status"].mode()[0],inplace=True) 
```

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True) 
```

```python
dataset["Children"].fillna(round(dataset["Children"].mean()),inplace=True)
```

```python
dataset["Commute Distance"].fillna(round(dataset["Commute Distance"].mean()),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset["Gender"].mode()[0]
```

```python
dataset
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(sparse=False)

dataset_encoded = pd.DataFrame (encoder.fit_transform(dataset[['Education']]))

dataset_encoded.columns = encoder.get_feature_names(['Education'])

dataset.drop(['Education'] ,axis=1, inplace=True)

dataset1 = pd.concat([dataset_encoded, dataset ], axis=1)
```

```python
dataset1
```

```python
encoder=OneHotEncoder(sparse=False)

dataset_encoded = pd.DataFrame (encoder.fit_transform(dataset1[['Occupation']]))

dataset_encoded.columns = encoder.get_feature_names(['Occupation'])

dataset1.drop(['Occupation'] ,axis=1, inplace=True)

dataset2 = pd.concat([dataset_encoded, dataset1], axis=1)
```

```python
dataset2
```

```python
encoder=OneHotEncoder(sparse=False)

dataset_encoded = pd.DataFrame (encoder.fit_transform(dataset2[['Region']]))

dataset_encoded.columns = encoder.get_feature_names(['Region'])

dataset2.drop(['Region'] ,axis=1, inplace=True)

dataset3 = pd.concat([dataset_encoded, dataset2], axis=1)
```

```python
dataset3
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset3["Marital Status"] = le.fit_transform(dataset3["Marital Status"])
dataset3["Gender"] = le.fit_transform(dataset3["Gender"])
dataset3["Home Owner"] = le.fit_transform(dataset3["Home Owner"])
dataset3["Bike Buyer"] = le.fit_transform(dataset3["Bike Buyer"])
```

```python
dataset3.columns
```

```python
dataset3
```

```python
dataset3.drop('ID',axis='columns', inplace=True)
```

```python
dataset3.columns
```

```python
dataset3.shape
```

```python
x = dataset3.iloc[:,0:21].values
y = dataset3.iloc[:,21].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
ypred = log.predict(x_test)
```

```python
ypred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ypred,y_test)
```
