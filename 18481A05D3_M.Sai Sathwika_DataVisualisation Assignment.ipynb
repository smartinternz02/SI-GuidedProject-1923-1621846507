# Graph Inside a Graph

```python
xpoints=np.array([5,8,2,4])
ypoints=np.array([4,6,8,3])
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])
ax1.plot(x,y)
ax2.plot(xpoints,ypoints)
```

# Seaborn

```python
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset("planets")
```

```python
df.head(5)
```

```python
df.plot(x="mass",y="distance",kind="pie")
```

```python
sns.relplot(data=df,x="mass",y="distance",hue="year")
```

```python
sns.relplot(data=df,x="mass",y="distance",col="year",col_wrap=2)
```

# distplot

```python
sns.distplot(df.mass,kde="True")
```

```python
sns.barplot(data=df,x="orbital_period",y="year")
```

# Countplot

```python
sns.countplot(data=df,x="year")
```

# Joint plot

```python
sns.jointplot(data=df,x="orbital_period",y="year")
```

# Box_plots

```python
sns.boxplot(x="distance",y="orbital_period",data=df)
```
