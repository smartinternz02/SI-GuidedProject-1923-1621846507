```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(sparse=False)

dataset_encoded = pd.DataFrame (encoder.fit_transform(dataset[['job']]))

dataset_encoded.columns = encoder.get_feature_names(['job'])

dataset.drop(['job'] ,axis=1, inplace=True)

dataset1 = pd.concat([dataset_encoded, dataset], axis=1)
```

```python
encoder=OneHotEncoder(sparse=False)

dataset_encoded = pd.DataFrame (encoder.fit_transform(dataset1[['marital']]))

dataset_encoded.columns = encoder.get_feature_names(['marital'])

dataset1.drop(['marital'] ,axis=1, inplace=True)

dataset2 = pd.concat([dataset_encoded, dataset1], axis=1)
```

```python
encoder=OneHotEncoder(sparse=False)

dataset_encoded = pd.DataFrame (encoder.fit_transform(dataset2[['education']]))

dataset_encoded.columns = encoder.get_feature_names(['education'])

dataset2.drop(['education'] ,axis=1, inplace=True)

dataset3 = pd.concat([dataset_encoded, dataset2], axis=1)
```

```python
encoder=OneHotEncoder(sparse=False)

dataset_encoded = pd.DataFrame (encoder.fit_transform(dataset3[['contact']]))

dataset_encoded.columns = encoder.get_feature_names(['contact'])

dataset3.drop(['contact'] ,axis=1, inplace=True)

dataset4 = pd.concat([dataset_encoded, dataset3], axis=1)
```

```python
dataset4
```

```python
encoder=OneHotEncoder(sparse=False)

dataset_encoded = pd.DataFrame (encoder.fit_transform(dataset4[['poutcome']]))

dataset_encoded.columns = encoder.get_feature_names(['poutcome'])

dataset4.drop(['poutcome'] ,axis=1, inplace=True)

dataset5 = pd.concat([dataset_encoded, dataset4], axis=1)
```

```python
dataset5
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset5["deposit"] = le.fit_transform(dataset5["deposit"])
dataset5["housing"] = le.fit_transform(dataset5["housing"])
dataset5["default"] = le.fit_transform(dataset5["default"])
dataset5["loan"] = le.fit_transform(dataset5["loan"])
```

```python
dataset5
```

```python
dataset5.drop(['day'] ,axis=1, inplace=True)
dataset5.drop(['month'] ,axis=1, inplace=True)
```

```python
dataset5
```

```python
dataset5.shape
```

```python
x = dataset5.iloc[:,0:35].values
y = dataset5.iloc[:,35].values
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
ypred = log.predict(x_test)
```

```python
ypred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ypred,y_test)
```
